```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("final_table.csv")
```

```python
data.head()
```

# Exploratory Data Analysis

```python
data.info()
```

```python
data.dtypes
```

```python
plt.figure(figsize=(6,5),dpi=75)
plt.hist(data["brand"],bins=15)
plt.xticks(rotation=45)
plt.show()
```

```python
data.brand.value_counts(normalize=True).plot.pie()
plt.show()
```

Solex is the most famous brand from above all

```python
data.job_industry_category.value_counts(normalize=True).plot.pie()
plt.show()
```

```python
data.job_industry_category.value_counts(normalize=True).plot.barh()
plt.show()
```

Manufacturing is at the top of the list

```python
corr=data.corr()
sns.heatmap(corr,annot=True)
```

```python
sns.countplot(data["gender"])
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

There are slightly more women than men in this data set. 

```python
sns.countplot(data["online_order"])
plt.show()
```

There are equal number of people using online and offline mode

```python
plt.hist(data["Age"],bins=50)
plt.title("Distribution of Age")
plt.xlabel("AGE")
plt.show()
```

There are lots of customer whose age are between 40 to 50

```python
plt.hist(x="Age",data=data[data["gender"] == "Male"],label="MALE",alpha=0.5)
plt.hist(x="Age",data=data[data["gender"] == "Female"],label="FEMALE",alpha=0.5)
plt.title('Distribution of Age by gender')
plt.xlabel('Age')
plt.legend()
plt.show()
```

There are a Men aged between 40 to 50 more than women aged between 40 to 50.
